```python
import requests
from bs4 import BeautifulSoup

def scrape_omim_disease_info(disease_name, url):

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    print(soup)

    # Assuming each section is under an 'h3' tag followed by paragraphs 'p'
    headers = soup.find_all('h3')
    for header in headers:
        section_title = header.get_text(strip=True)
        content = header.find_next_sibling('p').get_text(strip=True)

        file_name = f"{disease_name}_{section_title}.txt"
        with open(f'scraper_test/{file_name}', 'w') as file:
            file.write(content)
            print(f"Saved section '{section_title}' to '{file_name}'.")

# Usage
scrape_omim_disease_info('Leigh_Syndrome', 'https://www.omim.org/entry/256000')
```

```python
import io
import tokenizer
import os 

dir = "sample_files/"
all_files = os.listdir(dir)

import nltk
from nltk.tokenize import word_tokenize

# Tokenizing the text
nltk.download('punkt')


def count_tokens(filename):
    # Reading the text file
    with open(filename, 'r') as file:
        text = file.read()

    tokens = word_tokenize(text)

    # Counting the tokens
    number_of_tokens = len(tokens)

    return number_of_tokens

for file in all_files:
    print(f'{file}: {count_tokens(f"{dir}{file}")}')
```
